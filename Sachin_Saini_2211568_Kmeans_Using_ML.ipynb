# Finding optimum number of cluster using Elbow Method

```python
Centroid_list=[]
no_of_clusters=range(1,11)

for i in no_of_clusters:
  kmeans=cluster.KMeans(n_clusters=i, init="k-means++", max_iter=300, n_init=15, random_state=100)
  kmeans.fit(df)
  Centroid_list.append(kmeans.inertia_)


plt.plot(no_of_clusters,Centroid_list)
plt.title("Elbow Method")
plt.show()
```

We can conclude after k=2 the error starts decreasing so k=2.

# Making KMeans model

```python
kmeans=cluster.KMeans(n_clusters=2, init="k-means++", max_iter=300, n_init=15, random_state=100)
label= kmeans.fit_predict(df)
print(label)
```

```python
label_0=df[label==0]
label_1 = df[label == 1]

```

```python
plt.figure(figsize=(12,10))
cols = df.columns

plt.scatter(label_0[cols[0]], label_0[cols[1]], color = 'red',label="breast_cancer-malignant")
plt.scatter(label_1[cols[0]], label_1[cols[1]], color = 'black', label="breast_cancer-beningn")
plt.legend(loc='upper right')
plt.show()
```

```python
print(silhouette_score(df,kmeans.labels_))
```
